# Exercise 2 - Vectorized Logistic Regression

```python
# Define parameters and fictional X_t and y_t to test the Vectorized Logistic Regression:

theta_t = np.array([-2, -1, 1, 2])
X_t = np.hstack((np.ones(5).reshape(5,1),np.linspace(1,15,15).reshape(5,3, order='F')/10))
y_t = np.array([1,0,1,0,1])
lambda_t = 3
```

```python
# Define sigmoid function 
def sigmoid(x: np.array)-> np.array:
    '''
    Outputs the Sigmoid for each element of an array
    
    Args:
        x(np.array): Array with original values
    Returns:
        (np.array): Resulting sigmoid values of original matrix
    '''
    return(1 / (1 + np.exp(-x)))
```

We have to implement a regularized implementation of the Cost Function and the Gradient. This is similar to what we have did in former lessons:
   
<img src="assets/VectorizedLogisticRegression.JPG">
    

```python
def lrCostFunction(
    theta: np.array, 
    X: np.array, 
    y:np.array, 
    lambdas: int
) -> (np.array, np.array):
    '''
    Regularized logistic regression implementation 
    Returns Cost Function and respective gradient
    
    Args:
        theta(np.array): Theta parameter;
        X(np.array): Feature vector;
        y(np.array): output vector with each class;
        lambdas(int): regularization parameter;
    
    '''
    m = len(y)
    n = X.shape[1]
    
    #Hypothesis calculation
    h = sigmoid(np.dot(X,theta))
    
    a = -y*np.log(h)
    b = (1-y)*np.log(1-h)
    
    regul = lambdas/(2*m)
    
    # Compute cost function
    J = (sum(a-b))*(1/m) + regul * sum(theta[1:]**2)
               
    grad_regul = np.zeros(n)
    
    grad_regul[1:] = (lambdas/m)*theta[1:]
    
    # Compute gradient
    grad = ((1/m)*(np.dot(X.T,(h-y))) + grad_regul)
    
    return J, grad
```

```python
J, grad = lrCostFunction(theta_t, X_t, y_t, lambda_t)
```

```python
print('Cost for test example: {} '.format(J))
print('Gradients for test example: {} '.format(grad))
```

# Exercise 3 - One vs. All - Training

We need to optimize our Logistic Regression for each possible value of our outcome space. 
By developing a Logistic Regression for each outcome we can better capture the feature space that represents the drawing of each digit. Notice that we are only generalizing the functions we developed in previous lessons to more outcomes - this is, training more than one logistic regression at a time.

```python
# Define lambda for regularization
lambdas = 0.1 
```

```python
# Write one vs. all Classification function 

def oneVsAll(
    X: np.array, 
    y: np.array, 
    num_labels: int, 
    lambdas: float
) -> np.array:
    '''
    Function to develop regularized logistic regression for each target
    in outcome space. 
   
    Args:
        X(np.array): Feature Space;
        y(np.array): target array;
        num_labels(int): number of distinct outcomes;
        lambdas(float): regularization parameter;
    
    Returns:
        all_theta(np.array): Outputs Thetas for each model for each digit
    '''
    
    m = X.shape[0]
    n = X.shape[1]
    
    #Add intercept
    ones_vector = np.ones((len(X),1))
    X_extended = np.hstack((ones_vector,X))
    
    #Create placeholder for each theta
    all_theta = np.zeros((num_labels,n+1))
    
    #Options for optimization process
    options= {'maxiter': 100}
    
    #Create initial Theta
    init_theta = np.zeros((m,1))
    
    #Loop through each label and optimize the cost function for each one - Rewrite the Placeholder with the optimal theta found by the optimizer
    for label in np.arange(1,num_labels):
        init_theta = np.zeros((n+1,1))
        target = np.array([1 if val == True else 0 for val in y==label])
        res = optimize.minimize(lrCostFunction,
                        init_theta,
                        (X_extended, target, lambdas),
                        jac=True,
                        method='TNC',
                        options=options)
    
        all_theta[label-1,:] = res.x.T
    return all_theta
```

```python
all_theta = oneVsAll(X, y, num_labels, lambdas)
```

```python
def predictOnevsAll(
    all_theta: np.array, 
    X: np.array
)-> np.array:
    '''
    Predict one vs. all and output probability of digit for each outcome (0 to 9)  
    
    Args:
        all_theta(np.array): theta for each logistic regression trained.
        X(np.array): Each example features.
    Returns:
        np.array: Probability for each example of belonging to a specific class
    '''
    m = X.shape[0]

    #Add intercept
    ones_vector = np.ones((len(X),1))
    X_extended = np.hstack((ones_vector,X))
    
    return sigmoid(np.dot(X_extended, all_theta.T))
```

```python
# Get sigmoid predicted Vector
predicted = predictOnevsAll(all_theta, X)
```

```python
# Assign max probability to the predicted digit - Remember that 10 = 0 
result = (np.argmax(predicted, axis=1)+1).reshape(predicted.shape[0],1)
```

```python
print("Accuracy is {} ".format(sum(np.array([1 if val == True else 0 for val in result==y]))/len(y)))
```

We achieved an accuracy of 91.62%. For each 10 digits, we got ~9.1 of them correct. And we did it we a linear algorithm! It's definitely a remarkable milestone.
