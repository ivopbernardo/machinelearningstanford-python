# Exercise 1 - Support Vector Machines

```python
# Import numpy libraries to deal with matrixes and vectors
import numpy as np
# Import pandas do read data files
import pandas as pd
# Import matplotlib to plot data
import matplotlib.pyplot as plt

# Import math 
import math 

# Import scipy optimization function
from scipy import optimize, io
from scipy.ndimage import rotate

from sklearn.svm import LinearSVC
from sklearn.svm import SVC

# Matplotlib notebook property
%matplotlib inline
```

```python
# Use scipy Io to load matrix object with exercise data
ex6_file = io.loadmat('ex6data1.mat')
X = np.array(ex6_file['X'])
y = np.array(ex6_file['y'])
```

```python
def plotData(
    X: np.array,
    y: np.array
) -> None:
    '''
    Plots 2D plot with target shown as o and + 
    markers.
    
    Args:
        X(np.array): Array of features.
        y(np.array): Target value of each observations.
    Returns:
        None        
    '''
    plt.scatter(X[(y==0).ravel(),0], X[(y==0).ravel(),1], marker='o', facecolors='yellow', edgecolors='black')
    plt.scatter(X[(y==1).ravel(),0], X[(y==1).ravel(),1], marker='+', c='black')
```

```python
plotData(X.reshape(len(X),2),y)
```

Notice the outlier at **X=~0.01 and y=~4.2**. This outlier will play a huge role in understanding the SVM (particularly the parameter C) behavior.

```python
#Defining parameter C 
C = 1

def svmTrain(
    X: np.array, 
    y: np.array, 
    C: int,  
    max_iter: int
) -> None:
    '''
    Training Support Vector Classifier using sklearn SVM training.
    The implementation in sklearn converges slower than the octave version so we raise the number of iterations.
    
    Args:
        X(np.array): Original array of features.
        y(np.array): Target values of X observations.
        C(int): Regularization Parameters.
        max_iter(int): Number of iterations.
    Returns:
        w(np.array): Weights for variables.
        b(np.array): Intercept.
    '''
    
    svm_classifier = LinearSVC(C=C, max_iter = max_iter)
    svm_classifier.fit(X,y.reshape(len(y),))     
    
    # Get the coefficients (weights per variable and intercept) of the thetas trained by the Support Vector Machine
    w = svm_classifier.coef_  
    b = svm_classifier.intercept_  
    
    return w,b
```

```python
def visualizeBoundaryLinear(
    X: np.array, 
    y: np.array, 
    w: np.array, 
    b: np.array
) -> None:
    '''
    Maps decision boundary trained by the Support Vector Machine
    
    Args:
        X(np.array): Original array of features.
        y(np.array): Target values.
        w(np.array): Weight of variables optimized by SVM.
        b(np.array): Intercept value.
    Returns:
        None
    '''
    
    xp = np.linspace(min(X[:,0]), max(X[:,0]),100)
    yp = - (w[0,0]*xp + b)/w[0,1]
    
    plotData(X, y)
    plt.plot(xp,yp)
    
```

```python
# Grabbing weights and bias (intercept) from SVM 
w,b = svmTrain(X,y,C,100000)
```

```python
visualizeBoundaryLinear(X, y, w, b)
```

With a low C we relax the decision boundary enabling the algorithm to misclassify the outlier. 
<br>
**What happens when we set a high C?**

```python
C = 100
#Grabbing weights and bias (intercept) from SVM 
w,b = svmTrain(X,y,C,100000)
visualizeBoundaryLinear(X, y, w, b)
```

Notice how our outlier is now being correctly classified due to a higher C parameter. A higher C is equivalent to less regularization (if we compare this with a Linear Regression, it is the inverse of the Lambda parameter) being applied in the algorithm thus carrying a **risk of potential overfitting.**

# Exercise 2 - Implementing Gaussian Kernel

We have evaluated a linear decision boundary - a similar kernel to logistic regression. The SVM algorithm enables the construction of hyperplanes with other types of boundaries (for example, non-linear). 

The example that we are going to evaluate is the Gaussian Kernel which we will implement below:

<img src="assets/GaussianKernel.JPG">

```python
x1 = np.array([[1,2,1]])
x2 = np.array([[0,4,-1]])
```

```python
def gaussianKernel(
    x1: np.array, 
    x2: np.array, 
    sigma: int
) -> np.array:
    '''
    Evaluates gaussian kernel using features x1 and x2
    with sigma value.
    
    x1(np.array): Array of feature 1.
    x2(np.array): Array of feature 2.
    sigma(int): Sigma parameter.
    
    Return:
        sim(np.array): Gaussian Kernel Evaluation
    '''
    sim = math.exp((-(x1-x2)**2).sum()/(2*sigma**2))
    return sim
```

```python
sigma = 2
sim = gaussianKernel(x1,x2,sigma)

print('Gaussian Kernel between [1,2,1] and [0,4,-1] with sigma = {} is {}'.format(sigma,sim))
```

Two further visualize our Gaussian Kernel, we will load the second dataset of dummy data:

```python
# Use scipy Io to load matrix object with exercise data
ex6_file_2 = io.loadmat('ex6data2.mat')
X = np.array(ex6_file_2['X'])
y = np.array(ex6_file_2['y'])
```

```python
plotData(X.reshape(len(X),2),y)
```

```python
def svmTrainGaussian(
    X: np.array, 
    y: np.array, 
    C: int, 
    sigma: float
)-> SVC:
    '''
    We will use the sklearn built-in classification model
    with a gaussian kernel.
    
    The function trains and returns the SVC classifier.
    
    Args:
        X(np.array): Array of original features.
        y(np.array): Target values.
        C(int): Regularization parameter.
        sigma(float): Gaussian sigma parameter.

    Returns:
        svm_classifier(SVC): SVC trained classifier
    '''
    svm_classifier = SVC(kernel = 'rbf', C=C, gamma= np.power(sigma,-2.))
    svm_classifier.fit(X,y.reshape(len(y),))     
    
    return svm_classifier
```

```python
C = 1
sigma = 0.1
# Grabbing model from function svm trained object
model = svmTrainGaussian(X,y,C, sigma)
```

```python
def visualizeBoundary(
    X: np.array, 
    y: np.array, 
    model:SVC
) -> None:
    """
    Uses trained model and plots non-lineary decision boundary over grid
    Adapted grid using user @dibgerge code.
    
    Args:
        X(np.array): Original features.
        y(np.array): Target values 
        model(SVC): Trained Support Vector Classifier.
    Returns:
        None
    """
    plotData(X, y)

    # Generate Grid of 2-D Linear Space
    x1plot = np.linspace(min(X[:, 0]), max(X[:, 0]), 100)
    x2plot = np.linspace(min(X[:, 1]), max(X[:, 1]), 100)
    X1, X2 = np.meshgrid(x1plot, x2plot)

    # Iterate over grid and predict on new_X
    vals = np.zeros(X1.shape)
    for i in range(X1.shape[1]):
        new_X = np.stack((X1[:, i], X2[:, i]), axis=1)
        vals[:, i] = model.predict(new_X)

    plt.contour(X1, X2, vals, colors='y', linewidths=2)
    plt.pcolormesh(X1, X2, vals, cmap='YlGnBu', alpha=0.25, edgecolors='None', lw=0)
    plt.grid(False)
```

Let's visualize the decision boundary using a sigma of 1:

```python
visualizeBoundary(X,y,model)
```

The gaussian kernel was able to capture the non-linear effects o the y-variable related to X. 
<br>
